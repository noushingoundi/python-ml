#**Numpy**

#Create array

```python
import numpy as np

a = np.array([4,8,9,3,6])
b = np.arange(1,6)
c = np.zeros((4,3))
d = np.ones((2,3))


print("Array a:",a)
print("Array b:",b)
print("4x3:",c)
print("2x3:",d)
```

#Indexing & Slicing

```python
print("First element of a:",a[0])
print("Last two elements of a:",a[-2])
```

#Basic arithmetic operations

```python
print("a+2:",a+2)
print("a+2:",a+2)

```

BAsic statistics

```python
print("Square of a:",np.square(a))
print("Mean of a:",np.mean(a))
print("Standard deviation:",np.std(a))
```

Reshapping array

```python
matrix = np.arange(1,10).reshape((3,3))
print("Reshaped 3x3 matrix:")
matrix
```

Transpose

```python
print("Transpose:")
matrix.T
```

Boolean indexing

```python
print("Elements > 5:",matrix[matrix > 5])
```

#**Pandas**

#Create a DataFrame manually

```python
import pandas as pd
data = {
    'Name':['Alice','bob','charlie','david','eve'],
    'Age':[25,30,35,40,22],
    'Department':['sales','Engineering','HR','Engineering','Teacher'],
    'Salary':[50000,80000,40000,90000,45000]
}
df = pd.DataFrame(data)
df
```

Basic operations

```python
vprint("First 3 rows:")
df.head(3)
```

```python
print("column names:")
df.columns
```

```python
print("Data types:")
df.dtypes
```

```python
print("Describe numerical columns:")
df.describe()
```

Filtering

```python
print("People with salary > 50000:")
df[df['Salary'] > 50000]
```

Sorting

```python
print("Sorted by age:")
df.sort_values(by='Age')
```

Grouping and Aggregation

```python
grouped = df.groupby('Department')['Salary'].mean()
print("Average Salary by Department:")
grouped
```

Add new columns

```python
df['Salary in lakhs'] = df['Salary'] /100000
df
```

Matplotlib:Building Blocks of visualizations

Line Plot

```python
import matplotlib.pyplot as plt
years = [2019,2020,2021,2022]
revenue = [100,120,180,200]

plt.plot(years,revenue,marker='o',linestyle = '--',color = 'purple')
plt.title("Company Revenue Over Years")
plt.xlabel("Year")
plt.ylabel("Revenue (in Crores)")
plt.xticks(years)
plt.grid(True)
plt.show()
```

Bar Chart

```python
departments = df['Department'].value_counts()
departments.plot(kind='bar',color = 'teal')
plt.title("Employee count by department")
plt.xlabel("Department")
plt.ylabel("count")
plt.show()
```

Histogram

```python
ages = df['Age']
plt.hist(ages,bins=5,color='violet',edgecolor = 'black')
plt.title("age distribution")
plt.xlabel("age")
plt.ylabel("frequency")
plt.show()
```

Pie Chart

```python
df.groupby('Department').size().plot(kind='pie',autopct='%1.1f%%')
plt.title("Department proportion")
plt.ylabel("")
plt.show()
```

**Seaborn:Statistical Data Visualization**

```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips.head()
```

```python
print("\nInfo about the tips dataset")
tips.info()

print("\nDescription of the tips dataset:")
tips.describe(include='all')
```

Pair plot

```python
sns.pairplot(tips,hue = "sex")
plt.show()
```

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")

plt.title("Correlation Matrix")
plt.show()
```

```python
sns.countplot(x='day',data=tips,palette='Set2',hue='day',legend=False)
plt.title("count of records by day")
plt.show()
```

Boxplot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("boxplot of total bill by day")
plt.show()
```
